```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
#calculate the average fare paid by passengers in different classes
avg_fare_by_class = titanic.groupby('class')['fare'].mean()
```

```python
#use the bar() function from the pyplot module of Matplotlib to create the bar graph
plt.bar(avg_fare_by_class.index, avg_fare_by_class.values)
plt.title('Average Fare by Class on the Titanic')
plt.xlabel('Class')
plt.ylabel('Average Fare')
plt.show()
```
