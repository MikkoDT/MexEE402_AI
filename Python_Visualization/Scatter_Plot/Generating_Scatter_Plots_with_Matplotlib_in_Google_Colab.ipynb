```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('city_data.csv')
```

```python
plt.scatter(data['pop_density'], data['num_parks'])
plt.title('Relationship between Population Density and Number of Parks')
plt.xlabel('Population Density')
plt.xlabel('Number of Parks')
plt.show()
```
