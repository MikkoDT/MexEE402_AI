```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from plotnine import ggplot, aes, stat_summary, ggtitle
```

```python
titanic = sns.load_dataset('titanic')
```

**Creating a Bar Plot with Plotnine**

```python
# Creating a bar plot with Plotnine:
# ggplot() initializes the plot, aes() maps 'class' (x-axis) and 'fare' (y-axis).
# stat_summary() calculates and displays the mean fare per class.
# ggtitle() adds a descriptive title.

(ggplot(titanic)
+ aes(x='class', y='fare')
+ stat_summary(fun_y = np.mean, geom = 'bar')
+ ggtitle('Average Fare by Class on the Titanic')
)
```

**Creating an Interactive Scatter Plot with Plotly**

```python
import plotly.express as px
```

```python
# Using Plotly to create an interactive scatter plot:
# scatter() defines the plot with dataset, x/y variables, point color, title, and axis labels.

fig = px.scatter(titanic,

x='age',
y='fare',
color='class',
title='Fare vs Age Scatter Plot on the Titanic',
labels={'age':'Age', 'fare':'Fare'})

fig.show()
```

**Creating a Line Chart with Altair**

```python
import altair as alt
```

```python
# Creating a line chart with Altair:
# Chart() initializes the plot, mark_line() sets it as a line chart,
# encode() maps variables, and interactive()/properties() add interactivity and a title.

alt.Chart(titanic).mark_line().encode(
    x='age',
    y='mean(fare)',
    color='class',
    tooltip=['class', 'age', 'fare']
).interactive().properties(
    title='Average Fare by Age on the Titanic'
)
```
