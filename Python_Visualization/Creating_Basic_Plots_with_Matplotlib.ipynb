```python
# Import the matplotlib library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the URL where the dataset is located
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
```

```python
# Use pandas to load the dataset from the URL into a DataFrame
df = pd.read_csv(url)
```

```python
# Create a line chart
plt.figure(figsize=(10, 6))
df['beer_servings'].plot()
plt.title('Line Chart of Beer Servings')
plt.show()
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
df['beer_servings'].value_counts().plot.bar()
plt.title('Bar Chart of Beer Servings')
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
df['beer_servings'].plot.hist(bins=10)
plt.title('Histogram of Beer Servings')
plt.show()
```
