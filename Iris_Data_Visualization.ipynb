# **Step 5: Visualizing the Dataset**


Use a pair plot to visualize relationships between features and uncover patterns in the dataset.





```python
# Create a pair plot of the measurements, color-coded by species

sns.pairplot(iris, hue='species')

plt.show()
```

**Pair plot** shows scatterplots and histograms for all features, with color-coding by species to highlight differences and separations between them.



*   **Histograms (Diagonal):** Show feature distributions; e.g., sepal_width ~ normal, petal_length/width ~ bimodal.

*   **Scatterplots (Off-diagonal):** Show feature relationships; petal_length vs. petal_width shows strong positive correlation.

*   **Species Separation:** Color-coding reveals Iris-setosa is clearly distinct; versicolor and virginica overlap.

*   **Outliers/Overlaps:** Some versicolor points overlap with virginica, hinting at misclassification.






# **Step 6: Creating a Heatmap**
**Heatmap** is another excellent visualization to understand the correlation between different numerical features in the dataset.


Correlation values range from -1 to +1


*   +1 â†’ perfect positive relationship (as one increases, the other increases)

*   -1 â†’ perfect negative relationship (as one increases, the other decreases)

*   0 â†’ no linear relationship

The heatmap colors make it easy to spot:
*   Darker/red tones â†’ strong positive correlation

*   Darker/blue tones â†’ strong negative correlation

*   Lighter/neutral â†’ weak or no correlation











```python
plt.figure(figsize=(10,7))

# Use only numeric columns for correlation
sns.heatmap(iris.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')

plt.title('Heatmap for the Dataset')
plt.show()
```

# **Step 7: Creating Distribution Plots**
Interpreting the Distribution Plots:


*   **Sepal Length:** Setosa has shorter sepals (4.5â€“5.5 cm), while Versicolor and Virginica are longer but overlap.

*   **Sepal Width:** Setosa has larger widths; Versicolor and
Virginica overlap heavily, making this feature less useful.

*   **Petal Length:** Clear separation â€” Setosa < Versicolor < Virginica. Strong distinguishing feature.

*   **Petal Width:** Similar to petal length, clear separation among species.










```python
for column in iris.columns[:-1]:  # excluding the 'species' column
    sns.displot(iris, x=column, hue='species', kde=True)
    plt.title(f'Distribution for {column}')
    plt.show()
```

# **Step 8: Creating a Box Plot**
**Box plots,** also known as box-and-whisker plots, are helpful for visualizing the variance in the data and identifying potential outliers.

**Outliers** are data points that differ significantly from the majority of the data.

ðŸ”¹ They are unusually high or low values compared to the rest of the dataset.

ðŸ”¹ Outliers can appear due to measurement errors, variability in the data, or rare events.

ðŸ”¹ In plots (like boxplots), outliers are often shown as individual points outside the "whiskers" (the typical range of the data).

```python
sns.boxplot(data=iris, x='species', y='sepal_length', palette="Set2")
plt.title('Box Plot of Sepal Length for Each Species')
plt.show()
```

Interpreting the Box Plot:


*   **Setosa:** Median â‰ˆ 5.0 cm; range 4.3â€“5.8 cm; no outliers.

*   **Versicolor:** Median â‰ˆ 5.9 cm; range 5.0â€“7.0 cm; some outliers present.

*   **Virginica:** Median â‰ˆ 6.5 cm; range 5.8â€“7.7 cm; few outliers present.










# **Step 9: Creating Interactive Plots**

```python
fig = px.scatter(iris, x='sepal_width', y='sepal_length', color='species')

fig.show()
```

Interpreting the Interactive Scatter Plot:

*   **Setosa (blue):** Clear positive correlation â€“ wider sepals â†’ longer sepals.

*   **Versicolor (orange):** More scattered, weak positive trend.

*   **Virginica (green):** Scattered, weak correlation like Versicolor.

*   **Interactive Advantage:** Hovering shows exact sepal length/width for deeper exploration.
